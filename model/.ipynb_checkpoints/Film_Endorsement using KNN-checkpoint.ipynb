```python
#saving the above matrix into the dataframe
movies_users_df = ratings_df.pivot(index = 'movieId', columns = 'userId', values = 'rating').fillna(0)
movies_users_df.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
matrix_movies = csr_matrix(movies_users_df.values)
```

```python
matrix_movies
```

```python

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric = 'cosine', algorithm = 'brute', n_neighbors = 20)
model.fit(matrix_movies)
```

```python
from fuzzywuzzy import process
```

```python
def recommender(movie_name):
    ida = process.extractOne(movie_name, movies_df['original_title'])
    print(ida)
```

```python
recommender('toy story')
```

```python
def recommendation(movie_name, data, n):
    index = process.extractOne(movie_name, movies_df['original_title'])[2]
    print('Movie Selected: ', movies_df['original_title'][index], 'Index: ', index)
    print('Searching for the recommeded movies---------')
    distance, indices = model.kneighbors(data[index], n_neighbors = n)
    print(distance, indices)
```

```python
recommendation('toy story', matrix_movies, 10)
```

```python
#Final recommendation of the movies
def final_recommendation(movie_name, data, n):
    index = process.extractOne(movie_name, movies_df['original_title'])[2]
    print('Movie Selected: ', movies_df['original_title'][index], 'Index: ', index)
    print('Searching for the recommeded movies---------')
    try:  
        distance, indices = model.kneighbors(data[index], n_neighbors = n)
        for i in indices:
            tit = movies_df['original_title'][i].where(i!=index)
            return tit
            print(tit)
    except KeyError:
        print("Please decrease the n_neighbors")
        
```

```python
#based on the user rating
#user input
user_movie_name = input('Enter the movie name: ')
final_recommendation(user_movie_name, matrix_movies, 9)
```
